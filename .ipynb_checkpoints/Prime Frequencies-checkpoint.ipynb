Check http://arxiv.org/pdf/1603.03720v1.pdf statistics on first 50 000 000 primes.

```python
import os
import numpy as np
import time
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

```python
freq = np.zeros(100).reshape(10,10)

totalpn = 0
lastdigit = None

start = time.time()
for fi in log_progress(range(1, 51, 1)):
    f = open('./data/primes' + str(fi) + '.txt') 
    for line in f.readlines()[2:]:
        primes = ' '.join(line.split()).split(' ')
        if totalpn == 0:
            lastdigit = 2
            primes = primes[1:]
            totalpn = 1
        for p in primes:
            freq[int(p) % 10][lastdigit] += 1
            lastdigit = int(p) % 10
            totalpn += 1
        
print '-----ttime {} sec'.format(time.time()-start)
```

```python
freq = freq / totalpn
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
im = plt.imshow(freq, cmap='rainbow', interpolation='none')
plt.colorbar(im, orientation='horizontal', fraction=0.046, pad=0.04)
for x, y in np.ndindex(freq.shape):
    plt.text(x-0.4, y+0.15, '{0:.3f}'.format(freq[x][y]))
plt.xticks(range(0, 10, 1))
plt.yticks(range(0, 10, 1))
plt.title(r'$d_1 \rightarrow d_2$', fontsize=25)
plt.ylabel(r'$d_1$', fontsize=20, rotation=0)
plt.xlabel('$d_2$', fontsize=20)
ax.yaxis.tick_right()
ax.yaxis.set_label_coords(-0.05, 0.5)
plt.show()
```

```python
lfreq = np.array([[4623042., 7429438., 7504612., 5442345.], 
                  [6010982., 4442562., 7043695., 7502896.], 
                  [6373981., 6755195., 4439355., 7431870.], 
                  [7991431., 6372941., 6012739., 4622916.]])

ltnp = lfreq.sum()
ltnp
```

```python
lfreq = lfreq.T/ltnp
```

```python
lfreq
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
im = plt.imshow(lfreq, cmap='rainbow', interpolation='none')
plt.colorbar(im, orientation='horizontal', fraction=0.046, pad=0.06)
for x, y in np.ndindex(lfreq.shape):
    plt.text(x-0.4, y+0.15, '{0:.3f}'.format(lfreq[x][y]))
plt.xticks(range(0, 4, 1), ['1','3','7','9'])
plt.yticks(range(0, 4, 1), ['1','3','7','9'])
plt.title(r'$d_1 \rightarrow d_2$', fontsize=25)
plt.ylabel(r'$d_1$', fontsize=20, rotation=0)
plt.xlabel('$d_2$', fontsize=20)
ax.yaxis.tick_right()
ax.yaxis.set_label_coords(-0.05, 0.5)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
im = plt.imshow(lfreq-freq, cmap='rainbow', interpolation='none')
plt.colorbar(im, orientation='horizontal', fraction=0.046, pad=0.06)
for x, y in np.ndindex(lfreq.shape):
    plt.text(x-0.4, y+0.15, '{0:.3f}'.format(lfreq[x][y]-freq[x][y]))
plt.xticks(range(0, 4, 1), ['1','3','7','9'])
plt.yticks(range(0, 4, 1), ['1','3','7','9'])
plt.title(r'$Diff$', fontsize=25)
plt.ylabel(r'$d_1$', fontsize=20, rotation=0)
plt.xlabel('$d_2$', fontsize=20)
ax.yaxis.tick_right()
ax.yaxis.set_label_coords(-0.05, 0.5)
plt.show()
```
